#### This basic python image reader program achieve the following
1. Steps to upload images from google drive
2. Connecting to google drive 
3. Using CV2, read the image using imread 
4. Using matplot imshow 
5. Converting the scale from BGR to RGB
6. Converting to grayscale

```
from google.colab import drive
drive.mount('/content/drive')
```

```

img_path='/content/drive/My Drive/ml_data/stanford dog dataset/n02089973_1030.jpg'
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = cv2.imread(img_path)
```

```
type(img)
img.shape
```

```
plt.imshow(img)
```

###### You see that image is displayed in some wierd blue color.
###### This is because matplotlib expects data in Red Green Blue
###### and OpenCV in Blue Green Red
###### and hence image swapped

```
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```
plt.imshow(rgb_img)
```

```
rgb_img.shape
```

####### Read the image as a grayscale image

```
grey_scale_img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
plt.imshow(grey_scale_img)
```

#### Display portion of the image

```
nscl_img=rgb_img[0:200,0:300]
plt.imshow(nscl_img)
```

```
nscl_img.shape
```

##### Portion of the image in only one channel

```
nscl_img_gscale=rgb_img[0:200,0:300,1]
plt.imshow(nscl_img_gscale)
nscl_img_gscale.shape
nscl_img_gscale.ndim
```

#### Note  that Default Color Mapping that imshow uses is veridiss. It needs to be changed to Gray Scale. Veridiss shows lighter value as yellow and darker value as Blue.

```
plt.imshow(grey_scale_img,cmap='gray')
```

```
grey_scale_img.shape
```

###Note the third dimension is not there for greycale image

#### Resizing the image

```
rsd_img= cv2.resize(rgb_img,(900,800)) #### Note resizing parameter is opposite to numpy array sizing. i.e first argument is width & second argument is height
plt.imshow(rsd_img)
```

```
flpd_img=cv2.flip(rgb_img,0)    # flipped the image around X Axis 
plt.imshow(flpd_img)
```

```
flpd_img=cv2.flip(rgb_img,1)    # flipped the image around Y Axis 
plt.imshow(flpd_img)
```

```
cv2.imwrite('flipped_n02089973_1030.jpg',flpd_img)
```
